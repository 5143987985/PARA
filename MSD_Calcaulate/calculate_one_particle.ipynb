```python
import numpy as np
import pandas as pd
```

```python
one_data = pd.read_excel('one_particle.xlsx')
one_data.columns = ['X', 'Y']
one_data.head()
```

```python
one_data['X'].astype(float)
one_data['Y'].astype(float)
```

```python
# from point[j] to point[0]
x = list(one_data['X'])
y = list(one_data['Y'])
N = len(one_data)
msd_t = []
for t in range(1, N):
    sum = 0
    for i in range(t):
        sum += (x[i]-x[i-1])**2 + (y[i]-y[i-1])**2
    msd = sum/t
    
    msd_t.append(msd)
print(msd_t)
```

```python
x = list(one_data['X'])
y = list(one_data['Y'])
N = len(one_data)
msd_m = []
for m in range(1, N-1):
    diff_x = []
    diff_y = []
    for i in range(0, N-m-1):
        diff_x.append(x[i+m] - x[i])
        diff_y.append(y[i+m] - y[i])
    
    diff_x = np.array(diff_x)
    diff_y = np.array(diff_y)
    diff_sq = diff_x**2 + diff_y**2
    msd = np.mean(diff_sq, axis=0)
    
    msd_m.append(msd)
print(msd_m)
```
