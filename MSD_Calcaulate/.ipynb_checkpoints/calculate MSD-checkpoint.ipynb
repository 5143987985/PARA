```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw_data = pd.read_csv('C:/Users/tjut_/Desktop/calcaulate/21-NiV-M-1.3-GFP_t0001_ORG_Displacement^2.csv', usecols=[0,1,6])
raw_data.columns=['X', 'Y', "TrackID"]
raw_data.head()
```

```python
raw_data = raw_data.dropna(axis=0)
raw_data['TrackID'].astype(int)
raw_data['X'].astype(float)
raw_data['Y'].astype(float)
raw_data.head()
```

```python
ans = raw_data.groupby(['TrackID'])
ans.head(5)
```

```python
MSD_list = []

for key, value in ans: # trackID, points
    # print(key)
    x = list(value['X'])
    y = list(value['Y'])
    N = len(value) # points number in each line
    # print(N)
    
    msd_t = []
    for t in range(1, N):
        sum = 0
        for i in range(t):
            sum += (x[i]-x[0])**2 + (y[i]-y[0])**2
        msd = sum/t
        msd_t.append(msd)
    
    MSD_list.append(msd_t)
    
```
